# bernoulli's distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

```python
def bernoulli(x,p):
    if x==1:
        return p
    elif x==0:
        return 1-p
    else:
        return 0
print(bernoulli(0,0.6))
print(bernoulli(1,0.6))
```

```python
np.mean(bernoulli(1,0.6))
```

```python
np.var(bernoulli(0,0.6))
```

```python
p=0.6
samples=bernoulli.rvs(p,size=10)
print(samples)
```

```python
bernoulli.cdf(1,p)
```

# poisson distribution

```python
from scipy.stats import poisson
mu=3

samples=poisson.rvs(mu,size=10)
print(samples)
```

```python
a=poisson.pmf(4,mu)
print(a)
```

# multinomial 

```python
from scipy.stats import multinomial
n=10
pval=[0.6,0.3] # prob of outcomes
print(multinomial.rvs(n,pval,size=5))
print(multinomial.pmf([5,3],2,pval))
```
