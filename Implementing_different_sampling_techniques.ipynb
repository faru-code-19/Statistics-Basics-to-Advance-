```python
# extracting only numerical columns and making a new dataframe from it 
pop_df=tips[['total_bill','tip','size']] # population
pop_df # new dataframe
```

# Systematic Sampling

N- population size (len(pop_df))

n- sample size desried

k- intervals from start N//n

random start- 0 to k-1

```python
N=len(pop_df) # pop_size
n=20 # sample_size
k=N//n # intervals
print(f'intervals- {k}')

np.random.seed(42)
starting_point=np.random.randint(0,k-1)
print(f'starting_point is - {starting_point}')

sample_df=pop_df.iloc[starting_point::k].head(n)
sample_df
```

# stratified sampling 
# groupby and .sample using lambda function

```python
strata='sex'
samp_df=tips.groupby(strata,group_keys=True,observed=True).apply(lambda x : x.sample(n=5,random_state=42))
print(samp_df)
```

```python
strata='sex'
samp_df=tips.groupby(strata,group_keys=False).apply(lambda x : x.sample(n=5,random_state=42))
print(samp_df)
```
