```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
t=sns.load_dataset('titanic')
ct=t.dropna()
```

```python
ct.info()
```

```python
plt.figure(figsize=(10,9))
sns.histplot(ct['age'],kde=True,bins=50)
```

```python
# survival distribution
sns.countplot(data=ct,x='survived',hue='survived')
```

```python
# fare distibution
sns.histplot(data=ct,x='fare',kde=True)
```

```python
# AGE distribution by survival

sns.kdeplot(data=ct,x='age',hue='survived',fill=True,common_norm=)
```

```python
sns.countplot(x='class',data=ct,hue='class')
```
