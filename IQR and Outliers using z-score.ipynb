```python
import random
n=50
coins=['H','T']
h_cnt=0
t_cnt=0
for i in range(1,n+1):
    res=random.choice(coins)
    print(res,end=' ')
    if res=="H":
        h_cnt+=1
    else:
        t_cnt+=1
print()
print('tail-',t_cnt)    
print('head-',h_cnt)
prob_heads=h_cnt/n
print('head_prob-',prob_heads)
print('tails_prob-',1-prob_heads)
```

# detecting outliers using z-score

```python
import numpy as np

def detect_outliers(data):
    outliers = []
    pos_t = 3
    neg_t = -3
    mean = np.mean(data)
    std = np.std(data)

    for i in data:
        z = (i - mean) / std
        #print(i, z)
        if z > pos_t or z < neg_t:
            outliers.append(i)

    print("Outliers:", outliers)

data = [
    48, 50, 52, 49, 51, 47, 53, 50, 49, 52, 51, 48, 50, 49, 52,
    51, 47, 53, 50, 49, 52, 51, 48, 50, 49, 52, 51, 47, 53,
    50, 49, 52, 51, 48, 50, 49, 52, 51, 47, 53, 50, 49, 52,
    51, 48, 50, 49, 52, 51, 47, 300, 450, 800, -200,-500
]

detect_outliers(data)
```

```python
IQR=Q3-Q1
print(IQR)
```
