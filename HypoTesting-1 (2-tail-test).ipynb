```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

#### Problem statement:
#### A restaurant owner asserts the average total bill for customers is â‚¹20.To validate this assertion, we will conduct a hypothesis test using a significance level of 0.05 and a population standard deviation of â‚¹8.34. A random sample of customer bills from the â€œtipsâ€ dataset will be analyzed.
### Null Hypothesis (Hâ‚€): Average total bill for customers is â‚¹20
# ð»0:ðœ‡=20
### Alternative Hypothesis (Hâ‚):Average total bill for customers is not â‚¹20
# ð»1:ðœ‡â‰ 20


```python
# step -1 : define the hypothesis
pop_mean=20
significance_level=0.05
pop_std=8.34
sample_size=30 
sample_mean=22 # hypoyhetical sample mean
```

```python
# step-2 : calculate standard error (SE)
SE=pop_std/np.sqrt(sample_size)
print(SE)
```

```python
# step - 3 : calculate Test statistics as n=30 , calculate z-score/statistics
z_score=(sample_mean-pop_mean)/SE
print(z_score)
```

```python
# step - 4 calculate p -value
p_val=2*(1-stats.norm.cdf(abs(z_score))) # for 2 tailed test
print(p_val)
```

```python
# step - 5: determine critical z-value
z_critical=stats.norm.ppf(1-(significance_level/2))
print(z_critical)
```

```python
if abs(z_score)>z_critical:
    print("Reject Null hypo")
else:
    print("Fail to Reject Null Hypo")
```

```python
print("Decision on P_val")
if p_val<significance_level:
    print("Reject Null hypo")
else:
    print("Fail to Reject Null Hypo")
```

# Plotting time

```python
# plotting the sampling distributions for 2 tailed test 
# Define x and y 
x=np.linspace(15,25,1000)
y=stats.norm.pdf(x,pop_mean,SE)

# plot x and y
plt.plot(x,y,color="blue",label="Sampling Distributions")

#plot population and sample mean vertically 
plt.axvline(pop_mean,color="black",linestyle='--',label="Population mean")
plt.axvline(sample_mean,color="green",linestyle='-',label="Sample mean")

# convert z critical value to x for 
x_right=pop_mean + z_critical * SE
x_left=pop_mean - z_critical * SE
plt.axvline(x_right,color='red',linestyle='-',label='+ve critical Value')
plt.axvline(x_left,color='red',linestyle='-',label='-ve critical Value')

# shading the red region
plt.fill_between(x,y,where=x<x_left,color="red",alpha=0.5)
plt.fill_between(x,y,where=x>x_right,color="red",alpha=0.5)


plt.title('Sampling Distributions of 2 Tail Tests')
plt.xlabel('Sample mean')
plt.ylabel('prob density')
plt.legend(loc="best")
plt.show()
```

# Final Interpretaion

### At the 5% significance level, we fail to reject the null hypothesis.There is insufficient evidence to conclude that the average total bill for customers is different from â‚¹20.
