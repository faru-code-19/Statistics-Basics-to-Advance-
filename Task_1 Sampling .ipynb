## 3) An ice-cream manufacturer wants to study the sales of ice-creams to decide the production policies of the company. 
## He collects the data for the number of ice-creams sold in a month from all the 30 franchises.
## Find the mean and median sales of all the franchises
## sales = [21, 93, 62, 76, 73, 20, 56, 95, 41, 36, 38, 13, 80, 88, 34, 18, 40, 11, 25, 29, 61, 23, 82, 10, 92, 69, 60, 87, 14, 91]



```python
import numpy as n
sales = [21, 93, 62, 76, 73, 20, 56, 95, 41, 36, 38, 13, 80, 88, 34, 18, 40, 11, 25, 29, 61, 23, 82, 10, 92, 69, 60, 87, 14, 91]
mean=np.mean(sales)
print('mean-',mean)
median=np.median(sales)
print('median-',median)
```

## 4) Use the data given in the previous question to calculate the standard deviation of ice-cream sales.
## Also find the range in which the middle 50% of the sales would lie.

```python
std=np.std(sales)
print('std-',std)
Q1=np.percentile(sales,25)
Q3=np.percentile(sales,75)
print('middle 50% of sales-(IQR)',(Q1,Q3))
```

## 5) Again using the data of the previous question calculate the measure of skewness and kurtosis of ice-cream sales

```python
import scipy.stats as stats
skewness=stats.skew(sales)
kurtosis=stats.kurtosis(sales)
print('skew-',skewness)
print('kurtosis-',kurtosis)
```

### 6) An ice-cream manufacturer wants to study the sales of ice-creams to decide the production policies of the company. He collects the data for the number of ice-creams sold in a month from all the 30 franchises. Additionally he also has data for the average work hours of each franchise last month. Is there a relationship between the sales and working hours? Comment on the nature of relationship.

```python
sales = [21, 93, 62, 76, 73, 20, 56, 95, 41, 36, 38, 13, 80, 88, 34, 18, 40, 11, 25, 29, 61, 23, 82, 10, 92, 69, 60, 87, 14, 91]
hours=[7, 9.5, 8, 8.5, 9, 6.5, 7, 10.5, 9, 7.5, 8, 6, 10, 8.5, 6, 6.5, 7, 5, 6.5, 6.5, 7.5, 6.5, 8, 8.5, 9.5, 10,7.5, 8.5, 6.5, 10.5]
correlation=np.corrcoef(sales, hours)[0,1]
print(correlation)

#  -- Yes,relationship between the sales and working hours, as working hour increases sales also increases.
#  -- Its a positive correlation
```

## 9) The odds that it will rain tomorrow is 15:10. What is the probability that it will not rain tomorrow


```python
#9
success=15   
failure=10   
P_rain=success/(success+failure)
P_not_rain=1-P_rain
print("Probability that it will rain:",P_rain)
print("Probability that it will NOT rain:",P_not_rain)
```
