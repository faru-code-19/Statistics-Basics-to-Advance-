### 3. Perform a One-Sample T-Test
### Use this dataset of sample weights of protein bars (ideal weight is 50g):
### weights = [49.5, 50.1, 50.3, 49.8, 50.0, 49.6, 49.9, 50.2, 50.4, 49.7]
### Test whether the mean weight is significantly different from 50g using a one-sample T-Test.


```python
# Answer-3
```

```python
import scipy.stats as stats
weights = [49.5, 50.1, 50.3, 49.8, 50.0, 49.6, 49.9, 50.2, 50.4, 49.7]
# Null hypothesis (H₀) : The mean weight is 50 g
# Alternative hypothesis (H₁) : The mean weight is different from 50 g 
# 2 tail test

sample_size=10
t_stats,p_val=stats.ttest_1samp(weights,50)
if p_val<=0.05:
    print("reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```

#### 4. Perform an Independent Two-Sample T-Test
#### Given the scores of students in two different teaching methods:
#### method_A = [78, 85, 82, 88, 91, 76, 83]
#### method_B = [72, 79, 77, 74, 73, 78, 75]
#### Use an independent T-Test to check if there's a significant difference in the means

```python
# Answer-4
```

```python
# Ho: Mean score of Method A = Mean score of Method B
# H1: Mean score of Method A != Mean score of Method B

import scipy.stats as stats
method_A = [78, 85, 82, 88, 91, 76, 83]
method_B = [72, 79, 77, 74, 73, 78, 75]
t_stat, p_v = stats.ttest_ind(method_A, method_B)
if p_v<=0.05:
    print("reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```

#### 6. Perform a One-Way ANOVA Test

#### You have exam scores from students in three different classrooms:

#### class_1 = [85, 88, 90, 92, 86]
#### class_2 = [78, 81, 79, 77, 80]
#### class_3 = [90, 91, 89, 88, 92]
#### Use one-way ANOVA to check if the mean scores across the classes are significantly different.


```python
# Answer-6
```

```python
#### Ho: Mean scores of all classes are equal
#### H1: At least one class mean is different

import scipy.stats as stats

class_1 = [85, 88, 90, 92, 86]
class_2 = [78, 81, 79, 77, 80]
class_3 = [90, 91, 89, 88, 92]

f_stat, p_valu = stats.f_oneway(class_1, class_2, class_3)
if p_valu<=0.05:
    print("reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```

#### 8. Perform a Chi-Square Test of Independence
#### Given the following contingency table of gender vs. preferred mode of learning:

#### Sex Online Offline
#### Male	20      30
#### Female  35      15

#### Perform a Chi-Square test to check whether gender is associated with learning preference.


```python
# Answer-8
```

```python
# Null Hypo - variables are independent
# Alternatice Hypo - variables are not independent
import numpy as np
import scipy.stats as stats
data=np.array([[20,30],[35,15]])
chi_stats,p_va,_,_=stats.chi2_contingency(data)
if p_va<=0.05:
    print("reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
    print("variables are independent")

```
