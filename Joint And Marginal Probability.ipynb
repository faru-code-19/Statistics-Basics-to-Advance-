```python
import numpy as np
import pandas as pd
import seaborn as sns
data = {
    'Outlook': ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Rain', 'Rain', 'Overcast', 'Sunny', 'Sunny', 'Rain', 'Sunny', 'Overcast', 'Overcast', 'Rain'],
    'Temperature': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild'],
    'Humidity': ['High', 'High', 'High', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'High'],
    'Wind': ['Weak', 'Strong', 'Weak', 'Weak', 'Weak', 'Strong', 'Strong', 'Weak', 'Weak', 'Weak', 'Strong', 'Strong', 'Weak', 'Strong'],
    'Play Tennis': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']
}

df=pd.DataFrame(data)
df
```

```python
# contingencty table 
contingency_table=pd.crosstab(df['Outlook'],df['Play Tennis'])
contingency_table
```

```python
# convert freq to proportions 
# marginal prob
ct_margins=pd.crosstab(df['Outlook'],df['Play Tennis'],margins=True,normalize=True)
ct_margins
```

```python
# joint prob
jp=pd.crosstab(df['Outlook'],df['Play Tennis'],normalize=True)
jp
```

```python
tips=sns.load_dataset('tips')
tips.head(10)
```

```python
# contingency table
ct=pd.crosstab(tips['sex'],tips['smoker'])
ct
```

```python
#joint probability
jp=pd.crosstab(tips['sex'],tips['smoker'],normalize=True)
jp
```

```python
# marginal probability
mp=pd.crosstab(tips['sex'],tips['smoker'],margins=True,normalize=True)
mp
```

```python
# conditional prob
cp=pd.crosstab(tips['sex'],tips['smoker'],normalize='index')
cp
```

```python
cp=pd.crosstab(tips['sex'],tips['smoker'],normalize='columns')
cp
```
