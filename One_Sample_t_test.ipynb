# One sample t-Test

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
# tips dataset
tips=sns.load_dataset('tips')
tips
```

```python
#total bill 
#Is the average total bill significantly different from ₹20?
# (Ho) - average total bill significantly different from ₹20
# H1 - average total bill is not significantly different from ₹20
```

```python
alpha=0.05
```

```python
pop_mean=20
pop = tips['total_bill']   # population-like data
sample_size = 26

sample = np.random.choice(pop, sample_size, replace=False)
print(sample)
sample_mean=np.mean(sample)
```

```python
t_stats,p_val=stats.ttest_1samp(sample,pop_mean)
print(t_stats)
print(p_val)
```

```python
if p_val<=alpha:
    print("Reject Ho")
else:
    print("Fail to reject Ho")
```

## There is insuuficient info for average total bill is not significantly different from ₹20
