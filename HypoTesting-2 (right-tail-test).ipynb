#### Problem statement:
#### A restaurant owner asserts the average total bill for customers is greater than ‚Çπ20.To validate this assertion, we will conduct a hypothesis test using a significance level of 0.05 and a population standard deviation of ‚Çπ8.34. A random sample of customer bills from the ‚Äútips‚Äù dataset will be analyzed.
### Null Hypothesis (H‚ÇÄ): Average total bill for customers <= ‚Çπ20
# ùêª0:ùúá <= 20
### Alternative Hypothesis (H‚Çê):Average total bill for customers is > ‚Çπ20
# ùêª1:ùúá > 20

# Right Tailed TEst


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
# Step 1: Define hypotheses and parameters
pop_mean = 20
significance_level = 0.05
pop_std = 8.34
sample_size = 30
sample_mean = 22  # hypothetical sample mean
print('sample mean-',sample_mean)

# Step 2: Calculate Standard Error (SE)
SE = pop_std / np.sqrt(sample_size)
print("Standard Error:", SE)

# Step 3: Calculate Z-score (test statistic)
z_score = (sample_mean - pop_mean) / SE
print("Z-score:", z_score)

# Step 4: Calculate p-value (1- right tailed test)
p_val = (1 - stats.norm.cdf(z_score))
print("P-value:", p_val)

# Step 5: Calculate critical Z-value
z_critical = stats.norm.ppf(1 - significance_level)
print("Z-critical:", z_critical)

# Step 6: Decision using critical value
if (z_score) > z_critical:
    print("Reject Null Hypothesis (Z-critical method)")
else:
    print("Fail to Reject Null Hypothesis (Z-critical method)")

# Step 7: Decision using p-value
if p_val < significance_level:
    print("Reject Null Hypothesis (p-value method)")
else:
    print("Fail to Reject Null Hypothesis (p-value method)")
```

# Plot Timing

```python
# plotting the sampling distributions for 2 tailed test 
# Define x and y 
x=np.linspace(15,25,1000)
y=stats.norm.pdf(x,pop_mean,SE)

# plot x and y
plt.plot(x,y,color="blue",label="Sampling Distributions")

#plot population and sample mean vertically 
plt.axvline(pop_mean,color="black",linestyle='--',label="Population mean")
plt.axvline(sample_mean,color="green",linestyle='-',label="Sample mean")

# convert z critical value to x for 
x_right=pop_mean + z_critical * SE
plt.axvline(x_right,color='red',linestyle='-',label='+ve critical Value')

# shading the red region
plt.fill_between(x,0,y,where=x>x_right,color="red",alpha=0.5)

plt.title('Sampling Distributions of right Tailed Test')
plt.xlabel('Sample mean')
plt.ylabel('prob density')
plt.legend(loc="best")
plt.show()
```

# Final Interpreation

#### Based on the sample data, at the 5% significance level, we fail to reject the null hypothesis. 
#### There is insufficient evidence to support the claim that the average total bill for customers is greater than ‚Çπ20.
